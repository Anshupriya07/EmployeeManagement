```python
import os

hr_dept_list = ["Recruitment", "C&B", "Admin"]
hr_role_list = ["Member", "Lead", "Manager"]

def add_employee(emp_id, emp_name, emp_DOJ, emp_designation, emp_salary = 0):
    if not os.path.isfile('empdetails.txt'):
        open('empdetails.txt','a').close()
        log = open('login.txt','w')
        log.write('admin admin\n')
        log.close()
    emp_file = open('empdetails.txt','r+')
    emp_details = emp_file.readlines()
    for detail in emp_details:
        if emp_id == detail.split(',')[0]:
            print('The Employee Already Exists\n')
            return
    emp_file.write(emp_id+','+emp_name+','+emp_DOJ+','+emp_designation+','+str(emp_salary)+'\n')
    emp_file.close()
    log = open('login.txt','a')
    log.write(emp_id+" "+emp_name.split()[0]+'\n')
    log.close()
    
def add_hr(emp_id, hr_dept, hr_role):
    if not os.path.isfile('hrdetails.txt'):
        open('hrdetails.txt','a').close()
    hr_file = open('hrdetails.txt','r+')
    hr_details = hr_file.readlines()
    for detail in hr_details:
        if emp_id == detail.split(',')[0]:
            print('The Employee Already Exists\n')
            return
    
    if hr_dept in hr_dept_list and hr_role in hr_role_list:
        hr_file.write(emp_id+','+hr_dept+','+hr_role+'\n')
    else:
        print('Invalid HR Department or Role')
    hr_file.close()

def search_employee(emp_id):
    dets = []
    if not os.path.isfile('empdetails.txt'):
        print('No Employee Exists')
        return
    emp_file = open('empdetails.txt','r+')
    emp_details = emp_file.readlines()
    for detail in emp_details:
        if emp_id == detail.split(',')[0]:
            dets = (detail.split(','))
            print('Employee Details-:\n'+'Employee-ID: '+dets[0]+'\nEmployee Name: '+dets[1]+'\nDate of Joining: '+dets[2]+'\nDesignation: '+dets[3]+'\nSalary: '+str(dets[4]))
            break
    emp_file.close()
    if len(dets) == 0:
        print('Employee Not Found. Recheck ID')

def search_hr(emp_id):
    dets = []
    if not os.path.isfile('hrdetails.txt'):
        print('No HR Exists')
        return
    hr_file = open('hrdetails.txt','r+')
    hr_details = hr_file.readlines()
    for detail in hr_details:
        if emp_id == detail.split(',')[0]:
            dets = (detail.split(','))
            print('HR Details-:\n'+'Employee-ID: '+dets[0]+'\nDepartment: '+dets[1]+'\nDesignation: '+dets[2])
            break
    hr_file.close()
    if len(dets) == 0:
        print('HR Not Found. Recheck ID')

def upd_employee(emp_id):
    dets = []
    if not os.path.isfile('hrdetails.txt'):
        print('No Employee Exists')
        return
    emp_file = open('empdetails.txt','r+')
    emp_details = emp_file.readlines()
    for i in range(len(emp_details)):
        if emp_id == emp_details[i].split(',')[0]:
            dets = (emp_details[i].split(','))
            print('Employee Details-:'+'\nDesignation: '+dets[3]+'\nSalary: '+str(dets[4]))
            emp_details.pop(i)
            break
    emp_file.close()
    if len(dets) == 0:
        print('Employee Not Found. Recheck ID')
        return
    new_name = input('Enter New Name (N/n if Unchanged) -')
    new_doj = input('Enter New Date of Joining (N/n if Unchanged) -')
    new_desig = input('Enter New Designation (N/n if Unchanged) -')
    new_sal = input('Enter New Salary (N/n if Unchanged) -')

    if new_name == 'N' or new_name == 'n': 
        new_name = dets[1]
    if new_doj == 'N' or new_doj == 'n': 
        new_doj = dets[2]
    if new_desig == 'N' or new_desig == 'n': 
        new_desig = dets[3]
    if new_sal == 'N' or new_sal == 'n': 
        new_sal = dets[4]

    emp_file = open('empdetails.txt','w')
    emp_details.insert(i,emp_id+','+new_name+','+new_doj+','+new_desig+','+new_sal+'\n')
    emp_file.writelines(emp_details)

def remove_employee(emp_id):
    if not os.path.isfile('empdetails.txt'):
        print('No Employee Exists.')
        return
    emp_file = open('empdetails.txt','r')
    emp_details = emp_file.readlines()
    emp_file.close()
    for i in range(len(emp_details)):
        if emp_id == emp_details[i].split(',')[0]:
            emp_details.pop(i)
            emp_file = open('empdetails.txt','w')
            emp_file.writelines(emp_details)
            emp_file.close()
            log_file = open('login.txt','r')
            log_details = log_file.readlines()
            log_file.close()
            for i in range(1,len(log_details)):
                if emp_id == log_details[i].split()[0]:
                    log_details.pop(i)
                    log_file = open('login.txt','w')
                    log_file.writelines(log_details)
                    log_file.close()
                    return
    print("Employee Not Found. Recheck ID")

def remove_hr(emp_id):
    if os.path.isfile('hrdetails.txt'):
        hr_file = open('hrdetails.txt','r')
        hr_details = hr_file.readlines()
        hr_file.close()
        for i in range(len(hr_details)):
            if emp_id == hr_details[i].split(',')[0]:
                hr_details.pop(i)
                hr_file = open('hrdetails.txt','w')
                hr_file.writelines(hr_details)
                hr_file.close()
                return
    else:
        print('No HR Exists')
        return

    if os.path.isfile('empdetails.txt'):
        emp_file = open('empdetails.txt','r')
        emp_details = emp_file.readlines()
        emp_file.close()
        for i in range(len(emp_details)):
            if emp_id == emp_details[i].split(',')[0]:
                emp_details.pop(i)
                emp_file = open('empdetails.txt','w')
                emp_file.writelines(emp_details)
                emp_file.close()
                log_file = open('login.txt','r')
                log_details = log_file.readlines()
                log_file.close()
                for i in range(1,len(log_details)):
                    if emp_id == log_details[i].split()[0]:
                        log_details.pop(i)
                        log_file = open('login.txt','w')
                        log_file.writelines(log_details)
                        log_file.close()
                        return
    else:
        print('No HR Exists')
        return
    print("Employee Not Found. Recheck ID")

def admin_module():
    print('1: Add Employee\n2: Add HR\n3: View Employee Details\n4: View HR Details\n5: Update Employee Details\n6: Remove Employee\n7: Remove HR\nExit(Any Character)')
    while 1:
        choice = input('Enter Choice: ')
        if(choice == '1'):
            emp_id = input('Enter Employee ID: ')
            emp_name = input('Enter Employee Name: ')
            emp_doj = input('Enter Employee Date of Joining: ')
            emp_desig = input('Enter Employee Designation: ')
            emp_sal = input('Enter Employee Salary: ')
            print()
            add_employee(emp_id, emp_name, emp_doj, emp_desig, emp_sal)
        elif(choice == '2'):
            emp_id = input('Enter HR ID: ')
            hr_dept = input('Enter HR Department: ')
            hr_role = input('Enter HR Role: ')
            print()
            add_hr(emp_id,hr_dept,hr_role)
        elif(choice == '3'):
            emp_id = input('Enter Employee ID: ')
            print()
            search_employee(emp_id)
        elif(choice == '4'):
            emp_id = input('Enter HR ID: ')  
            print()
            search_hr(emp_id)
        elif(choice == '5'):
            emp_id = input('Enter Employee ID: ')
            print()
            upd_employee(emp_id)
        elif(choice == '6'):
            emp_id = input('Enter Employee ID: ')
            print()
            remove_employee(emp_id)
        elif(choice == '7'):
            emp_id = input('Enter Employee ID: ')
            print()
            remove_hr(emp_id)
        
        else :
            print('Exiting Portal...')
            break
        
def employee_module(emp_id):
    if os.path.isfile('hrdetails.txt') and os.path.isfile('empdetails.txt'):
        hr_file = open("hrdetails.txt",'r')
        emp_file = open("empdetails.txt",'r')
        list_emp = emp_file.readlines()
        list_emp_id = [i.split(',')[0] for i in list_emp]
        list_hr = hr_file.readlines()
        list_hr_id = [i.split(',')[0] for i in list_hr]
        if(emp_id in list_emp_id and emp_id not in list_hr_id):
            print("Welcome {}!!".format(list_emp[list_emp_id.index(emp_id)].split(',')[1]))
            while 1:
                print("1 - View Details\n2 - View HR\nq - Exit")
                c= input("Enter your choice: ")
                if c == '1':
                    print(list_emp[list_emp_id.index(emp_id)])
                elif c == '2':
                    for i in list_hr_id:
                        print(list_emp[list_emp_id.index(i)])
                elif c == 'q':
                    break
            #show(employee_menu)
        elif(emp_id in list_emp_id and emp_id in list_hr_id):
            print("Welcome {} from HR!!".format(list_emp[list_emp_id.index(emp_id)].split(',')[1]))
            while 1:
                print("1 - View Details\n2 - View Employees\nq - Exit")
                c= input("Enter your choice: ")
                if c == '1':
                    print(list_emp[list_emp_id.index(emp_id)])
                elif c == '2':
                    z = input("Enter Designation: ")
                    for i in list_emp_id:
                        if z in list_emp[list_emp_id.index(i)]:
                            print(list_emp[list_emp_id.index(i)])
                        elif z == 'All':
                            print(list_emp[list_emp_id.index(i)])
                elif c == 'q':
                    break
    else:
        print('Files not found')
def validate():
    file = open("login.txt")
    list_emp = file.readlines()
    list_id = [i.split(' ')[0] for i in list_emp]
    while 1:
        username = input('Enter Username: ')
        password = input('Enter Password: ')
        if username == password == 'admin':
            admin_module()
        elif username in list_id:
            password_file = list_emp[list_id.index(username)].split(' ')[1]
            password_processed = password.split('\\')[0]
            #print(username,password_processed == password)
            if(password_processed == password):
                employee_module(username)
            else:
                print("Invalid Password")
                break
            break
        else:
            print('Invalid User Type. Exiting.')
            break

validate()
```
